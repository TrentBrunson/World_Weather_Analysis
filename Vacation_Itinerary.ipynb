```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests # Import the requests library.

# Import the API key.
from config import g_key
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
import gmaps
import gmaps.datasets

WeatherPy_vacay_df = pd.read_csv('weather_data/WeatherPy_vacation.csv')

WeatherPy_vacay_df
```

```python
# reset index to City
vacay_new_df = WeatherPy_vacay_df.set_index("City")
vacay_new_df
```

```python
# From the vacation search map, choose four cities that a customer might want to visit. 
# They should be close together and on the same continent.
# Matara, Wattega, Hambantota, Batticaloa in LK (Sri Lanka)

# Create separate DataFrames for each city on the travel route, using the loc attribute.
matara = vacay_new_df.loc[['Matara']]
matara
```

```python
wattegama = vacay_new_df.loc[['Wattegama']]
wattegama
```

```python
hambantota = vacay_new_df.loc[['Hambantota']]
hambantota
```

```python
batticaloa = vacay_new_df.loc[['Batticaloa']]

batticaloa
```

```python
four_cities = pd.concat([matara, hambantota,  wattegama, batticaloa])
four_cities
```

```python
four_cities = four_cities.reset_index().set_index('City_ID')

four_cities
```

```python
city_coords = four_cities.loc[:,['City', 'Lat', 'Lng']]

city_coords
```

```python
city_lat_long = city_coords.to_numpy()
city_lat_long
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}<dd>{Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in four_cities.iterrows()]
```

```python
# Add a pop-up marker for each city.
locations = four_cities[["Lat", "Lng"]]
fig = gmaps.figure(center=(7.0, 81.0), zoom_level=7)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
